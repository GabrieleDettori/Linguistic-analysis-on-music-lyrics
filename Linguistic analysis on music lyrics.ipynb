## State of the art

Topic models are a group of statistical methods for classifying documents by the contents of their text. The general intuition behind these methods is that documents, or in this case song lyrics, can be classified based on the occurence and frequency of the words contained within them. 
We will print some wordclouds that highlights (depending on the size of the printed word) the most frequent words in each specific period of time analyzed. We will see that the use of the frequencies of words to understand their importance is however a limited tool, and that the Latent Dirichlet allocation is a more advanced and efficient approach that will detect a certain number of main topics and the words most associated with them. We can learn a bit more from the related Wikipedia page: 

_In natural language processing, the latent Dirichlet allocation (LDA) is a generative statistical model that allows sets of observations to be explained by unobserved groups that explain why some parts of the data are similar. For example, if observations are words collected into documents, it posits that each document is a mixture of a small number of topics and that each word's presence is attributable to one of the document's topics. LDA is an example of a topic model and belongs to the machine learning toolbox and in wider sense to the artificial intelligence toolbox._

## Topic modelling

### Creating an LDA model of the entire corpus to check if there are meaningful words releted to the most frequent 5 topics

Firstly i create two sets, filenames and raw_text (that contains all the raw text of our corpus).

```python
filenames = [] 
raw_texts = [] 


for pathname in os.listdir("."):
    if pathname.startswith("1") or pathname.startswith("2"):
        for file in os.listdir(pathname):
            if file.startswith("."):
                continue
            else:
                if file=="CAPAREZZA-VERITASUPPOSTE_03.txt" or file=="MARRACASH-MARRACASH_01.txt" or file=="SALMO-MIDNITE_03.txt" or file=="TEDUA-VITAVERA_01.txt" or file=="TEDUA-VITAVERA_02.txt" or file=="TEDUA-VITAVERA_14.txt":
                    continue
                else:
                    filenames.append(file)
                    filepath = os.path.join(pathname, file)
                    infile = open(filepath, "r", encoding = "utf8")
                    text = infile.read()
                    raw_texts.append(text)
                    infile.close()

    else:
        continue
        
        
```

Then I create a pipline of the raw_text. Again, in order to avoid encoding problems, I had to avoid unknown charcathers and NoneType files.

```python
nlp = stanza.Pipeline("it", processors = "tokenize,pos,lemma")
import sys
annotated_docs = []
for text in raw_texts:
    if not text=="Ã" and not text=="¬" and not type(text)=="NoneType":
        annotated_docs.append(nlp(text))
```

Then I filtered the text using the Italian stopword.

```python
oc_poses = set(["ADJ", "ADV", "INTJ", "NOUN", "PROPN", "VERB"])

filtered_docs = []

for doc in annotated_docs:
    filtdoc = []
    for w in doc.iter_words():
        if w is not None and w.lemma is not None:
            if w.lemma in stopwords_it or w.text in stopwords_it or not w.upos in oc_poses:
                continue
            else:
                filtdoc.append(w.lemma)
        
        filtered_docs.append(filtdoc)
```

Then I created dictionary and filtered it and created a bow_corpus. Then I saved the dictionary and the corpus on new files.

```python
print(f'The vocabulary size is {len(filtered_docs)}')
```

```python
dct = gensim.corpora.Dictionary(filtered_docs)
```

```python
dct.filter_extremes(keep_n=10000, keep_tokens=None)
```

```python
bow_corpus = []

for doc in filtered_docs:
    bow_corpus.append(dct.doc2bow(doc))
    
print(f'The vocabulary size decreases to {len(filtered_docs)}')
```

```python
gensim.corpora.Dictionary.save(dct, 'data/dct.dict')  # save dictionary to disk

gensim.corpora.MmCorpus.serialize('data/bow_corpus.mm', bow_corpus)  # save corpus to disk
```

```python
loaded_dct = gensim.corpora.Dictionary.load('data/dct.dict')  # load the dictionary

loaded_corpus = gensim.corpora.MmCorpus( 'data/bow_corpus.mm')  # load the corpus
```

```python
ldamodel = gensim.models.ldamulticore.LdaMulticore(
    loaded_corpus, num_topics=5, id2word = loaded_dct, passes= 15,
    random_state = 140)
```

Let's visualize the most frequent words for topic emerged from the LDA model.

```python
tavolozza = plt.cm.get_cmap('Set2').colors

num_top_words = 10

topic2top_words = dict()
topics_descriptions = dict(ldamodel.show_topics(formatted=False, num_words = num_top_words))
for topic, words in topics_descriptions.items():
    topic2top_words[topic] = [[], []]  # una lista formata da due liste
    for stats in words:
        topic2top_words[topic][0].append(stats[0])
        topic2top_words[topic][1].append(stats[1])


for topic_id in topic2top_words:
    plt.subplot(2, 3, topic_id + 1)
    plt.barh(topic2top_words[topic_id][0], 
             topic2top_words[topic_id][1],
             color = tavolozza[topic_id])

    plt.gca().invert_yaxis()
    plt.title(f'Topic #{topic_id}')

plt.tight_layout()
plt.show()
```

```python
loaded_dct1 = gensim.corpora.Dictionary.load('data/dct1.dict')  # load the dictionary

loaded_corpus1 = gensim.corpora.MmCorpus( 'data/bow_corpus1.mm')  # load the corpus

loaded_dct2 = gensim.corpora.Dictionary.load('data/dct2.dict')  # load the dictionary

loaded_corpus2 = gensim.corpora.MmCorpus( 'data/bow_corpus2.mm')  # load the corpus
```

Now it is time to create our LDA models and print the most relevant topics with the associated most relevant keywords.

```python
ldamodel1 = gensim.models.ldamulticore.LdaMulticore(
    loaded_corpus1, num_topics=5, id2word = loaded_dct1, passes= 20,
    random_state = 140)
```

```python
tavolozza = plt.cm.get_cmap('Set2').colors

num_top_words = 10

def showtopicwords (ldamodel):
    topic2top_words = dict()
    topics_descriptions = dict(ldamodel.show_topics(formatted=False, num_words = num_top_words))
    for topic, words in topics_descriptions.items():
        topic2top_words[topic] = [[], []]  # una lista formata da due liste
        for stats in words:
            topic2top_words[topic][0].append(stats[0])
            topic2top_words[topic][1].append(stats[1])


    for topic_id in topic2top_words:
        plt.subplot(2, 3, topic_id + 1)
        plt.barh(topic2top_words[topic_id][0], 
                 topic2top_words[topic_id][1],
                 color = tavolozza[topic_id])

        plt.gca().invert_yaxis()
        plt.title(f'Topic #{topic_id}')

    plt.tight_layout()
    plt.show()
    
print ("Main topics from 1990 to 2005")
showtopicwords(ldamodel1)
```

```python
ldamodel2 = gensim.models.ldamulticore.LdaMulticore(
    loaded_corpus2, num_topics=5, id2word = loaded_dct2, passes= 20,
    random_state = 140)
```

```python
print ("Main topics from 2005 to 2020")
showtopicwords(ldamodel2)
```

# Dicussion

## Few words about lexical richness

Analyzing the lexical richness value of the corpus it seems that there are not many differences between the different time frames analyzed. Unfortunately, however, we do not have corpus of very similar length and therefore this data is not very reliable. Looking at the most frequent words that emerged from wordclouds, perhaps we will be able to understand something more about the lexical knowledge of Italian rappers.

## Topics emerged from frequencies wordclouds

To make these wordclouds sensible I had to remove some verbs by inserting them in the blacklist. Verbs like "know", "go", "do", "stay" ... are used a lot by rappers in the  “story-telling” or “ego-trip” songs, where rappers tell their story in first person. These verbs were repeated in all periods of time, so I decided to remove them to be able to notice more significant differences and words more related to specific contexts.

Temporal analysis:

* From 1990 to 1995 it is interesting to see that the word “guaglione” emerged. It is a dialectic word used as slang style. We will see that in the next years the slang words will become more frequent but there won’t be any slang word from a dialect. Other frequent words like “soldo” and “strada” are frequent because of the social protest style of these years. As we said in the introduction in those years rap was closely linked to the themes of social centers and the Italian left.

* From 1995 to 2000 wee see more self-referential words like “rap”, “rima”, “testo”, “microfono”, “musica”,”stile"…  These words denote the growth of a characteristic slang of the genre. In a period when rap was not yet famous in Italy, there was a need to explain what rap was through the songs themselves.

* From 2000 to 2005 we see the frequent use of the term "calling" reflects the increase in the use of mobile phones that has conditioned the way of life of the whole society in the 2000s. 

* From 2005, the year of the first mainstream rap, to 2010; we see the bad language emerging. Also the word “pazzo”, which rimes with a very used Italian bad word, has emerged, and this can show us how much the importance of the rhyme can prevail over the other literal components of a lyric. We start to see words related to the love theme like “donna” (rappers were mostly male). 

* From 2010 to 2015 we see the first english word emerged (“flow”). I think that if we could check American lyrics of these years we will see that term flow was very popular also in the U.S. Another frequent word emerged is “fumo”, referred to the use of marijuana and hashish. Love themes remained popular (“cuore”, “godere”)

* From 2015 to 2020 we can see the interjections “ya”, typical of the “trap” style, came also in Italy. Bad words didn’t emerge and the theme of love is approached in a more direct way (“amare”).

I think that the information emerged from wordclouds is more referred to the sounds of the words than to the topic they refer too. For example, we talked about the social protest in rap. A word like “soldo” (money), can be used to talk about social inequities but also to talk about the successful lifestyle of a rapper. I think that the topic modeling tools, grouping the words in topics, will make us understand if the use of the word “soldo” has changed over the years.


## Topics emerged from LDA model

As we can see in the first list of topics that emerged (entire corpus), the first topic refers to the classic lexicon of rapping ("rhyme", "text", "rap", in other versions also the word "microphone" came out) and the most characteristic word of this topic it's just rhyme. Many rappers say that rhyme is the only true rule of the genre and this talk of rhymes in the lyrics confirms the great attention that rappers devote to this word.

The second argument, determined more by the word haste, in my opinion denotes the theme of negative emotions or in any case of discomfort. "Devil", "evening", "rotten", "fuck" are all words associated with the rapper's inner discomfort. This allows us to say a few words about a way of using rap that is very popular both in America (Eminem) and in Italy (Fabri Fibra): the use of rap as a self-analysis of one's own discomforts and traumas. Rappers say that writing about these things is cathartic for them, it helps them to overcome certain problems. A famous phrase by Fabri Fibra gives us a precise idea of this way of using rap: "Io scrivo per necessità, c'ho scosse emozionali, sfogandomi è la cura ai miei disturbi cerebrali". ("I write out of necessity, I have emotional shocks, venting out is the cure for my brain disorders").

The third topic that emerged I must admit is difficult to define, and in my humble opinion, wanting to find a meaning for it would seem a bit forced.

The fourth and fifth topic, on the other hand, can be a starting point for analysis relating to what we said previously on the word money and on the theme of social protest. In fact, in both topics the word money appears but if in the first topic it is associated with words such as: "business", "war", "conflict", "crazy", "defeat" and "road", in the second topic it is associated with themes more related to the rapper's personal history: "home", "coming back", "area" ... The information that the subsequent analysis gave us can help us in the analysis of the theme of money. When we divide the LDA models for pre-music industry rap and music industry rap, we see that the fourth topic, the one that seems closest to the theme of social protest, belongs to the pre-music industry rap, while the theme of money in the more recent rap lyrics is not associated to political or social protest terms. This data is interesting as it reflects what was said at the beginning about the birth of Italian hip-hop.
The topics that emerged from pre-music industry rap are the most classic and self-referential ones, as if the main topic of the rappers was rap. In my opinion, this is due to the need to explain hip-hop culture to Italians, a distant and different culture, for which Italians were not prepared. The most recent rap has fortunately freed itself from this need.

As we have seen, the stereotype of self-referential themes in rap has been confirmed. The theme of love is present but is not as central as in the analysis that Kshirsagar made on American rap. The theme of party never emerged in Italian rap. The stereotype of vulgar rap can be said to be true especially after rap has entered the record economy. We can't say that Italian rap talks about social protest more than American's one, because  since American rap has already been at the top of the music charts for many years, it is normal that niche themes of this type have not emerged from the analyzes. The theme of the Anglophone language emerged but very little, that is only with the word "flow" and "dj". I think that LDA model helped us elaborating social analysis more than simple frequencies wordcloud, so we can say that researchers from Wasada and Kyoto were right.

I think that the main topic modelling tools are still not suitable for poetic language, especially that of rap where rhyme has priority over everything. Perhaps a good computational analysis tool for rap lyrics should have an algorithm that gives less importance to words that rhyme. Moreover, the more popular rap becomes, the less important the lyrics become.
I wonder how a type of analysis like mine could be implied with other Music Information Retrieval tools. Combining the results obtained from these types of models with those of a more literary analysis could allow companies such as Spotify, Apple or Google music to carry out profiling operations that obtain results on both the musical and lyrical preferences of the users.

We begin to see that tools such as [Autotune](https://it.wikipedia.org/wiki/Auto-Tune) that improve the intonation of singers are spreading fast. Who knows if in a few years even tools that will allow you to write better lyrics or to find "flows" more suitable for rap beats will be fashionable? This question is incorporated into a series of very important arguments for the field of Digital Humanities, such as the question of whether artistic production can be reproduced by artificial intelligences.
